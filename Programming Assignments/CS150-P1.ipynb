```python
import random
class game:
    def __init__(self):
        self.n = 0
        pass
    def play(self):
        rint = random.randint(1,3)
        print("Which do you choose? (1 is a rock, 2 is a paper, 3 is a scissor)")
        print("1 - Rock")
        print("2 - Paper")
        print("3 - Scissor")
        inp2 = input("->")

        if inp2 == "1":
            if rint == 1:
                print("Tie")
                print(f"System chose a {rint}")
            if rint == 2:
                print("Loss")
                self.n -= 1
                print(f"System chose a {rint}")
            if rint == 3:
                print("Win")
                self.n+=1
                print(f"System chose a {rint}")
        elif inp2 == "2":
            if rint == 1:
                print("Win")
                self.n+=1
                print(f"System chose a {rint}")
            if rint == 2:
                print("Tie")
                print(f"System chose a {rint}")
            if rint == 3:
                print("Loss")
                self.n -= 1
                print(f"System chose a {rint}")
        elif inp2 == "3":
            if rint == 1:
                print("Loss")
                self.n -= 1
                print(f"System chose a {rint}")
            if rint == 2:
                print("Win")
                self.n+=1
                print(f"System chose a {rint}")
            if rint == 3:
                print("Tie")
                print(f"System chose a {rint}")
        else:
            self.n-=4
            print("Choose a number on the menu next time, dummy")
  
class Menu:
    def __init__(self):
        pass

    def printmenu(self):    
        c = True 
        su = 0
        sc = 0
        t =0
        while c == True:
            print("ROCK PAPER SCISSORS: Please select an option from the following")
            print("0 - Play")
            print("1 - Exit")
            inp = input("->")
            if inp == "0" or inp.lower() == "play":
                g = game()
                g.play()
                if g.n == -1:
                    sc +=1
                elif g.n == 1:
                    su +=1
                elif g.n == 0:
                    t +=1
                print(f"Score for User: {su} Computer: {sc} Tied: {t}")
            elif inp == "1" or inp.lower() == "exit":
                c = False
            else:
                print("Not a menu option")
m = Menu()
m.printmenu()
```
