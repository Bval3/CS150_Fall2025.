```python
plt.scatter(df['Quality of Sleep'], df['Stress Level'])
plt.title('Sleep Quality vs. Stress')
plt.xlabel('Quality of Sleep (1-10)')
plt.ylabel('Stress Level (1-10)')
plt.show()
```

```python
#The project specifically asks to use bar plots for hstrogams so I will assume that I will only use the plt.bar function
bar1 = df['Quality of Sleep'].value_counts().sort_index()
plt.title('Sleep Quality')
plt.xlabel('Quality of Sleep (1-10)')
plt.ylabel('Count')
plt.bar(bar1.index, bar1.values)
```

```python
bar1 = df['Stress Level'].value_counts().sort_index()
plt.bar(bar1.index, bar1.values)
plt.title('Stress Level')
plt.xlabel('Stress level (1-10)')
plt.ylabel('Count')
```

```python
occupations = ["Nurse","Doctor", "Software Engineer","Salesperson","Teacher"]
for job in occupations:
    subset = df[df["Occupation"] == job]
    counts = subset["Stress Level"].value_counts().sort_index()
    plt.bar(counts.index, counts.values)
    plt.xlabel("Stress Level")
    plt.ylabel("Count")
    plt.title(f"Stress Level Distribution for {job} roles")
    plt.xticks(range(1, 11))
    plt.show()
```

The statistics from the data produced relatively similar results with few notable differences. The mean values for each occupation category varied little. The highest reported sleep quality came from the Nurse and Doctor roles, and the lowest reported sleep quality came from the Software Engineer role. The highest reported stress levels came from the Nurse, Doctor, and Software Engineer roles, and the lowest reported stress levels came from the Nurse and Doctor roles. The highest standard deviation for sleep quality was observed in Software Engineers, and the lowest standard deviation was observed in the Salesperson role. The highest standard deviation for stress level was observed in Nurses, and the lowest standard deviation for stress level was observed in the Salesperson role.

The graph plots of the data provide more interesting insights. On the scatter plot of sleep quality vs. stress level, the two variables show a negative relationship, where stress level decreases as sleep quality increases. However, the strength of the correlation between these two variables is low-moderate, as their distribution on the scatter plot is somewhat dispersed. For the graphs of stress for each occupation category, the most notable distribution is the Salesperson role, as out of all the distributions, there is high uniformity among all individuals. The entire distribution is reported at stress level 7 (hence the previous 0 in standard deviation for stress levels). The Nursing role's distribution is also interesting, as while most occupations have a majority at either higher or lower reported stress levels with a left or right skew, the Nurse distribution is symmetric, with two nearly equal-sized portions at higher and lower stress levels.
